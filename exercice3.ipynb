```python
import pandas as pd

# Charger le dataset
pos_data = pd.read_csv("POS_CleanData.csv")
pos_data
```

```python
print("Statistiques descriptives :")
print(pos_data.describe(include='all'))
```

```python
close_up_data = pos_data[pos_data['Brand'] == 'Close-up']
print("\nLignes où la colonne Brand est 'Close-up' :")
print(close_up_data)
```

```python
print(len(pos_data))
```

```python
df = pos_data[(pos_data['Revenue($)'] >= 10000) & (pos_data['Revenue($)'] <= 15000)]
print("\nDataFrame où Revenue($) est entre 10 000 et 15 000 :")
print(df)
```

```python
high_revenue_low_traffic = pos_data[(pos_data['Revenue($)'] > 25000) & (pos_data['Page_traffic'] < 1000)]
print("\nProduits avec Revenue($) > 25 000 USD et Page_traffic < 1000 :")
print(high_revenue_low_traffic)
```

```python
pos_data['Date'] = pd.to_datetime(pos_data['Date'], errors='coerce')
print("\nColonne Date après conversion en date-heure :")
print(pos_data['Date'].head())
```

```python
# Spécifier le format de la date si connu, par exemple 'YYYY-MM-DD'
pos_data['Date'] = pd.to_datetime(pos_data['Date'], format='%Y-%m-%d', errors='coerce')
print("\nColonne Date après conversion en date-heure :")
print(pos_data['Date'].head())
```
